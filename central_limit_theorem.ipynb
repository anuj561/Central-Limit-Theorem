```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
import random
randomlist = []
for i in range(0,1000000):
  n = random.randint(1,300)
  randomlist.append(n)
```

```
frequency = []
for i in set(randomlist):
  frequency.append(randomlist.count(i))
freq=pd.DataFrame(set(randomlist),frequency).reset_index()
```

```
freq.columns=[['frequency','number']]
```

```
plt.plot(freq['number'],freq['frequency'])
```

```
a=np.reshape(np.array(randomlist),(40000,25))
```

```

mean_list = []
for i in a:
  sum=0
  for j in i:
    sum = sum + j
  mean_list.append(int(sum/25))
```

```
len(mean_list)
```

```
mean_frequency = []
for i in set(mean_list):
  mean_frequency.append(mean_list.count(i))
mean_freq=pd.DataFrame(set(mean_list),mean_frequency).reset_index()
```

```
mean_freq.columns=[['frequency','number']]
```

```
plt.plot(mean_freq['number'],mean_freq['frequency'])
```
